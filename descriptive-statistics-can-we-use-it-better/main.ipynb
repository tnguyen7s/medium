### Rents - Mean and Median

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
prices = pd.read_csv("rents.csv")
```

```python
fig, ax = plt.subplots()
ax.hist(prices["Price"]/1000, bins=100)
ax.set_xticks(range(0, 70, 3))
ax.set_xlabel("Thousand dollars", fontsize=14)
plt.show()
fig.savefig('rents.jpg',
            format='jpeg',
            dpi=120,
            bbox_inches='tight')
```

```python
prices.mean()
```

```python
prices.median()
```

### Canvas outcome - Percentages data

```python
data = pd.DataFrame([('Spring 2021', 97.28, 626), ('Summer 2021', 88.24, 51), ('Fall 2021', 93.82, 842), ('Spring 2022', 91.37, 2851)], 
                    columns=["Term", "Percentage of Student Outcomes Mastered", "Total of Student Outcomes"])
```

```python
data
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(data['Term'], data['Percentage of Student Outcomes Mastered'], color="#48A9B7")
ax2.set_xlabel('Term', fontsize=14)
ax2.set_ylim((80,100))
ax2.set_ylabel('Percentage of student outcomes mastered (%)', fontsize=12)

fig2.savefig('line.jpg',
            format='jpeg',
            dpi=120,
            bbox_inches='tight')
plt.show()
```

```python
# first axis
fig3,ax3 = plt.subplots()
ax3.bar(data['Term'], data['Total of Student Outcomes'], color="#B75648")
ax3.set_ylabel('Total of Student Outcomes', fontsize=14, color="#B75648")
ax3.set_xlabel('Term', fontsize=14)


# second axis
ax4=ax3.twinx()
ax4.plot(data['Term'], data['Percentage of Student Outcomes Mastered'], color="#48A9B7")
ax4.set_ylim((80,100))
ax4.set_ylabel('Percentage of student outcomes mastered (%)', fontsize=14, color="#48A9B7")

plt.show()
fig3.savefig('second_axis.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
